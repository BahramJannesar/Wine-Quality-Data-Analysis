```python
wine_df.info()
```

```python
wine_df.head()
```

```python
fig = plt.figure(figsize = (15,15))
ax = fig.gca()
wine_df.hist(ax=ax , color = 'g')
plt.style.use('bmh')
plt.show()
```

```python
fig , ax = plt.subplots(2,2)
fig.set_size_inches([15 ,15])
ax[0 , 0].scatter(white_wine['volatile acidity'] , white_wine['residual sugar'] , c = 'g')
plt.setp(ax[0, 0], xlabel='volatile acidity')
plt.setp(ax[0, 0], ylabel='residual sugar')
ax[0 , 0].set_title('distributions between volatile acidity and residual sugar')

ax[0 , 1].scatter(white_wine['pH'] , white_wine['alcohol'] , c = 'g')
plt.setp(ax[0, 1], xlabel='pH')
plt.setp(ax[0, 1], ylabel='alcohol')
ax[0 , 1].set_title('distributions between pH and alcohol')

ax[1 , 0].scatter(white_wine['volatile acidity'] , white_wine['alcohol'] , c = 'g')
plt.setp(ax[1, 0], xlabel='volatile acidity')
plt.setp(ax[1, 0], ylabel='alcohol')
ax[1 , 0].set_title('distributions between volatile acidity and alcohol')

ax[1 , 1].scatter(white_wine['pH'] , white_wine['residual sugar'] , c = 'g')
plt.setp(ax[1, 1], xlabel='pH')
plt.setp(ax[1, 1], ylabel='residual sugar')
ax[1 , 1].set_title('distributions between pH and residual sugar')
```

```python
wine_df.groupby('color')['quality'].mean().plot(kind ='bar' , color =['red' , 'pink'])
```

```python
sns.catplot(x ='color' , y = 'quality', data = wine_df, kind = 'box' , hue = 'color' , ci = 'sd')
plt.show()
```

```python
sns.catplot(x ='color' , y = 'quality', data = wine_df, kind = 'bar' , hue = 'color' , ci = 'sd' , estimator = np.mean)
plt.show()
```

```python
counts = wine_df.groupby(['quality' , 'color']).count()['pH']
```

```python
total = wine_df.groupby('color').count()['pH']
```

```python
proporations = counts / total
```

```python
proporations.plot(kind = 'bar'  , color = ['red' , 'white'] )
plt.xlabel('Color and Quality')
plt.ylabel('Proporation')
plt.title('Proporation by Color and Quality')
plt.show()
```
